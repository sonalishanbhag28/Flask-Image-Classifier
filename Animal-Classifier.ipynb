### Starting the steps:

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Convolution2D, MaxPooling2D, Flatten
```

```python
from tensorflow.keras.preprocessing .image import ImageDataGenerator
#three transformations for each image: shear_range, zoom_range, horizontal_flip
train_datagen = ImageDataGenerator (rescale= 1./255, shear_range= 0.2, zoom_range= 0.2, horizontal_flip=True)
test_datagen = ImageDataGenerator (rescale= 1./255)
```

```python
x_train = train_datagen.flow_from_directory(r'/content/Crop-animal data/trainset',target_size = (64,64),batch_size = 32,class_mode = "categorical")
x_test = test_datagen.flow_from_directory(r'/content/Crop-animal data/testset',target_size = (64,64),batch_size = 32,class_mode = "categorical")
```

```python
x_train.class_indices
```

```python
model= Sequential()
```

```python
model.add(Convolution2D(32,(3,3), input_shape=(64,64,3)))
```

```python
model.add(MaxPooling2D((2,2)))
```

```python
model.add(Flatten())
```

```python
model.add(Dense(units=128, kernel_initializer="random_uniform", activation= "relu"))
#add output layer
model.add(Dense(units=5, kernel_initializer="random_uniform", activation= "softmax"))
        
#compile the model
model.compile(optimizer="adam", loss="categorical_crossentropy" , metrics=["accuracy"])
```

```python
model.fit(x_train, steps_per_epoch=47, epochs=5, validation_data= x_test, validation_steps=20)
```

```python
model.save("animal.h5")
```
